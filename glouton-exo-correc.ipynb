  ##  LE PROBLEME DU SAC A DOS:
  


Un voleur s´introduit dans une maison avec un sac pouvant contenir au maximum 30kg. Il trouve 4 objets dont la masse, la valeur en euro et la densité de valeur sont indiquées dans le tableau ci-dessous.

  <img src="http://www.fredpeuriere.com/NSI/algorithmique-3/tab.png"/>


Un problème se pose à lui: <b>Quels objets faut-il mettre dans le sac de manière à optimiser la valeur totale sans dépasser la masse maximale autorisée pour le sac?</b>

<div class="alert alert-block alert-info">1) Exploration de toutes les solutions par <b>force brute</b>: Testez la fonction <b>combinaisons()</b> qui affiche l´ensemble des possibilités permettant de remplir le sac avec ces quatre objets (sans tenir compte de la contrainte sur la masse maximale) et qui retourne le nombre de combinaisons possibles. </div>

```python
def combinaisons():
    cpt=0
    for i in range (2):
        for j in range(2):
            for k in range (2):
                for l in range (2):
                    print(i,j,k,l)
                    cpt+=1
    return cpt

combinaisons()
```

<div class="alert alert-block alert-success">Remarque, il y a 2^4 possibilités, ce qui n´est pas énorme. Mais l´exploration par force brute devient trop coûteuse pour un grand nombre (n) d´objets: \begin{equation}2^n
\end{equation}</div>

<div class="alert alert-block alert-info">2) Modifiez la fonction combinaisons() de manière à afficher l´ensemble des possibilités permettant de remplir le sac (avec la masse correspondante) <b>qui ne dépassent pas les 30kg</b> et qui retourne le nombre de combinaisons possibles. </div>

<div class="alert alert-block alert-info">3) Modifiez la fonction combinaisons() de manière à afficher parmi les combinaisons possibles la valeur correspondante. Quelle est la <b>solution optimale?</b></div>

  <img src="http://www.fredpeuriere.com/NSI/algorithmique-3/tab.png"/>

```python
def combinaisons():
    cpt=0
    for i in range (2):
        for j in range(2):
            for k in range (2):
                for l in range (2):
                    masse=i*8+j*13+k*10+l*12
                    valeur=i*3+j*7+k*3+l*4
                    if masse<=30:
                        print(i,j,k,l,'masse:',valeur,'euros')
                        cpt+=1
    return cpt

combinaisons()
```

<div class="alert alert-block alert-info">6) Modifiez la fonction de tri par sélection qui permet de trier cette liste par <b>ordre décroissant des densités de valeur.</b></div>

```python
# la fonction trie une liste par ordre décroissant par séléction du maximum
def tri_Selection(liste):
    n=len(liste)
    for i in range(n-1):
        indice_min=i
        for j in range(i+1,n):
            if liste[j][3]>liste[indice_min][3]: #il suffisait de modifier cette ligne
                indice_min=j
        liste[i],liste[indice_min]=liste[indice_min],liste[i]
    return liste
tri_Selection(objets)
```

<b>Stratégie gloutonne:</b><br> On sélectionne les objets un à un par ordre de densités de valeur décroissantes et on ajoute l’objet sélectionné dans le sac si la masse maximale reste respectée.

<div class="alert alert-block alert-info">7) Complétez la fonction <b>sac(liste,masse_max)</b>. Elle prend en paramètre la liste d´objets et la masse maximale que le sac peut contenir. Elle doit retourner la solution gloutonne avec les objets à prendre et la valeur totale du butin en euros...</div>
Testez ensuite la fonction pour la liste d´objets triée et une masse maximale de 30kg.
 <img src="http://www.fredpeuriere.com/NSI/algorithmique-3/tab.png"/>

```python
def sac (liste,masse_max):
    n=len(liste)
    solution=[]
    valeur=0
    for i in range(n): 
        v=liste[i][2]
        if masse_max>=v:
            masse_max=masse_max-v # masse maximum restante
            valeur+=liste[i][1]
            solution.append(liste[i][0]) # on ajoute le nom des objets emportés
    return solution,valeur


print(sac(objets,30))
```

<div class="alert alert-block alert-info">8) La solution trouvée est-elle <b>optimale?</b>

<div class="alert alert-block alert-success">Non... le butin est de 10 euros alors que la solution optimale donne 11 euros. On retrouve le défaut de l´algorithme glouton: Il explorer les solution à partir de la plus grande valeur sans remettre en cause ce choix par la suite.</div>
